```python
train_dir = r'D:\8th SEMESTER\ML LAB\ML Lab Spring 25\train'
validation_dir = r'D:\8th SEMESTER\ML LAB\ML Lab Spring 25\validation'
test_dir = r'D:\8th SEMESTER\ML LAB\ML Lab Spring 25\test'
train_Cercospora_dir = os.path.join(train_dir, 'common_rust')
```

```python
datagen = ImageDataGenerator(
                            rotation_range=40,
                            width_shift_range=0,
                            height_shift_range=0,
                            shear_range=0,
                            zoom_range=0,
                            horizontal_flip=True,
                            fill_mode='nearest')
```

```python
datagen = ImageDataGenerator(
                             width_shift_range=0.2)
```

```python
from tensorflow.keras.preprocessing import image
fnames = [os.path.join(train_Cercospora_dir, fname) for fname in os.listdir(train_Cercospora_dir)]
img_path = fnames[3]
img = image.load_img(img_path, target_size=(150, 150))
x = image.img_to_array(img)
x = x.reshape((1,) + x.shape)
i=0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 1== 0:
        break
plt.show()
```
